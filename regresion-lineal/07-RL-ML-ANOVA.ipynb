```python
mi_primera_anova = ols('charges_Sklearn ~ age+sex+bmi+children+smoker+region', data=df).fit()
sm.stats.anova_lm(mi_primera_anova)
```

Si PR es >=0.05, nuetsras variables predictoras influyen en nuestra variable respuesta, por lo que vemos, que todas nuestras variables, influyen 'charges'.

```python
mi_primera_anova.summary()
```

>- Nos da un R2 de 0.774, es decir, nuestras variables predictoras explican en un 77.4% nuestra variable respuesta.
>- Podemos que ver que todas nuestras variables predictoras excepto sexo (masculino), tienen relación positiva, si aumentan, aumenta el valor de 'charges. 
>- Con p-value<0.05 tenemos edad, bmi, fumadores, rechazamos H0, nuestras variables predictoras tienen efecto en nuestra variable respuesta.
